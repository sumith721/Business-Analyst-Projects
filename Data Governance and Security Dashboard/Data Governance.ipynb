```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X) 
```

```python
k_values = range(2, 11)
```

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
silhouette_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_scaled)
    silhouette_score_k = silhouette_score(X_scaled, kmeans.labels_)
    silhouette_scores.append(silhouette_score_k)
```

```python
optimal_k = k_values[silhouette_scores.index(max(silhouette_scores))]
kmeans = KMeans(n_clusters=optimal_k)
kmeans.fit(X_scaled)
```

```python
country_clusters = kmeans.predict(X_scaled)
cluster_labels = kmeans.labels_
```

```python
for k in range(optimal_k):
    # Assign cluster labels to a variable named 'cluster_labels'
    cluster_ESGData = ESGData[cluster_labels == k]
    print(f"Cluster {k+1}:")
    print(cluster_ESGData[['Indicator Name', 'Indicator Code']].head())
    print(cluster_ESGData.describe())
```
